```python
import pandas as pd
```

```python
df1 = pd.read_csv('file1.csv')
df2 = pd.read_csv('file2.csv')
df3 = pd.read_csv('file3.csv')
df3
```

```python
shape1 = df1.shape
shape2 = df2.shape
shape3 = df3.shape

print(shape1, shape2, shape3)
```

```python
column_names = [df1.columns, df2.columns, df3.columns]
column_names
```

```python
#rearange df2 columns
df2 = df2[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
        'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
        'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
df2.columns
```

```python
#rearange df3 columns
df3 = df3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
        'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
        'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
df3.columns
```

```python
column_names = [df1.columns, df2.columns, df3.columns]
column_names
```

```python
#standardizing header names
df1.columns = df1.columns.str.lower()
df2.columns = df2.columns.str.lower()
df3.columns = df3.columns.str.lower()
column_names = [df1.columns, df2.columns, df3.columns]
column_names
```

```python
#standardizing "state" column
df1 = df1.rename(columns= {"st":"state"})
df2 = df2.rename(columns= {"st":"state"})
column_names = [df1.columns, df2.columns, df3.columns]
column_names
```

```python
#concatinating Dframes:
concat_df = pd.concat([df1,df2,df3])
concat_df
```

```python
#which data types are numeric and which categorical:
concat_df.dtypes
```

```python
#droping 'education' and 'number of open complaints' columns
concat_df = concat_df.drop(columns=['education', 'number of open complaints'])
concat_df
```

```python
'''Correct the values in the column customer lifetime value. 
They are given as a percent, so multiply them by 100 and change dtype to numerical type.'''

concat_df['customer lifetime value'] = concat_df['customer lifetime value'].fillna('0%')
concat_df['customer lifetime value']
```

```python
#dropping duplicates
concat_df = concat_df.drop_duplicates()
concat_df
```

```python
#Filter out the data for customers who have an income of 0 or less.
negative_income = concat_df[concat_df['income'] <= 0.0]
negative_income
```
